```python
import dgl
import numpy as np
import torch
import networkx as nx
import matplotlib.pyplot as plt

#dgl : deep graph library
g= dgl.graph(([0,0,0,0,0],[1,2,3,4,5]), num_nodes = 6)
#g= dgl.graph((torch.LongTensor([0,0,0,0,0]), torch.LongTensor([1,2,3,4,5])), num_nodes=6)

#OR we can write
a = dgl.graph(([0,0,0,0,0],[1,2,3,4,5]))
g_edges = g.edges()

print(g_edges)
```

```python
gx= dgl.to_networkx(g)
print(gx)
```

```python
gx= dgl.to_networkx(g)
nx.draw_networkx(gx)
```

```python
# Assigning Node and Edge Features to Graph
g.ndata['value'] = torch.randn(6,3)  # assign a 3 dimensional node feature vector for each node
g.edata['weight'] = torch.randn(5,4) # assign a 4 dimensional edge feature vector for each edge.

g.ndata['value2'] = torch.randn(6,4,2) # node and edge features in DGL can be multi-dimensional.

print(g.ndata['value'])
```

```python
num_nodes = g.num_nodes()
num_edges = g.num_edges()

#out degrees of the center node
center_node_degree_out = g.out_degrees(0)
center_node_degree_in = g.in_degrees(0)
print("num_nodes",num_nodes)
print("num_edges",num_edges)
print(center_node_degree_out)
print(center_node_degree_in)
```
